```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculate_standard_deviation(series):
    if series.dtype == object:
        series = series.map({"No": 0, "Yes": 1}).fillna(series)

    counts = series.value_counts()
    values = counts.index.to_numpy(dtype=float)
    freqs = counts.to_numpy(dtype=float)

    mean = np.sum(values * freqs) / np.sum(freqs)
    variance = np.sum(freqs * (values - mean) ** 2) / (np.sum(freqs)-1)
    return np.sqrt(variance)
```
